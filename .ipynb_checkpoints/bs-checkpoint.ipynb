```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import json
```

```python
final_list=[]
url = 'https://quotes.toscrape.com/page/1'
while True:
    print(f'Start scraping {url}')
    try:
        r = requests.get(url, timeout=10)
        if r.status_code == 200:
            html = bs(r.content, 'html.parser')
            quotes = html.select('div[class=quote]')
            for num, quote in enumerate(quotes, 1):
                print(f'\tstart scraping quote #{num} from {len(quotes)}')
                text = quote.select_one('span[class=text]').get_text()
                author = quote.select_one('span small[class=author]').get_text()
                tag = quote.select_one('div[class=tags]').select('a[class=tag]')
                tags =  [x.get_text() for x in tag ]
                temp_dict = {
                    'text': text,
                    'author': author,
                    'tags': tags
                }
                final_list.append(temp_dict)
                
            try:
                next_button = html.select_one('li[class=next]').select_one('a')['href']
                url = 'https://quotes.toscrape.com' + next_button
            except Exception as e:
                print(f'Error: {e}')
                break
        else:
            print(f'Error status code: {r.status_code}')
            break
    except requests.Timeout:
        print('Timeout')
        break
        
print(f'\nDONE. Scraped {len(final_list)} quotes')

```

```python
with open('quotes.json', 'w') as q:
    json.dump(final_list, q)
            
```
