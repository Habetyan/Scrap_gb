```python
import json
from pymongo import MongoClient
```

```python
client = MongoClient('mongodb://localhost:27017/')
```

```python
print(client.list_database_names())
```

```python
mdb = client.quotes
```

```python
with open('quotes.json', 'r') as js:
    qs = json.load(js)
```

```python
mdb.all_data.insert_many(qs)
```

```python
print(client.list_database_names())
```
